```python

import pandas as pd

flight_weather_data = pd.read_csv("/mnt/c/Users/tbrag/OneDrive/Desktop/bda602venv/ML Project/flight_weather_fulldata.csv")

# Count rows where both ori_TMIN and dest_TMIN are not NaN
count_valid = (flight_weather_data['ori_TMIN'].notna() & flight_weather_data['dest_TMIN'].notna()).sum()

print(f"Number of rows where both ori_TMIN and dest_TMIN are not NaN: {count_valid}")

#filter those rows out for models including weather
filtered_data = flight_weather_data[(flight_weather_data['ori_TMIN'].notna()) & (flight_weather_data['dest_TMIN'].notna())]

##export the new table##
filtered_data.to_csv('filtered_flight_weather.csv', index=False)
```

```python
# List of columns to keep
columns_to_keep = ['FL_DATE','FL_NUMBER','ori_TMIN','ori_TMAX','ori_SNOW','ori_SNWD','ori_AWND','ori_PRCP','dest_TMIN','dest_TMAX','dest_SNOW','dest_SNWD','dest_AWND','dest_PRCP']

# Drop all columns except the specified ones
filtered_flight_weather_edited = filtered_data.drop(columns=flight_weather_data.columns.difference(columns_to_keep))


##export the new table##
filtered_flight_weather_edited.to_csv('filtered_flight_weather_edited.csv', index=False)
```

```python
# Create binary classifier for delayed flights
condition = (
    (filtered_data['ARR_DELAY'] >= 15) |
    (filtered_data['DEP_DELAY'] >= 15) |
    (filtered_data['CANCELLED'] == 1)
)

# Assign the binary classification to a new column 'DELAYED'
filtered_data.loc[:, 'DELAYED'] = condition.astype(int)

# Display the first 5 rows and check DataFrame info
print(filtered_data.head(5))
print(filtered_data.info())
```

```python
# List of columns to keep
columns_to_keep_two = ['FL_DATE','AIRLINE','FL_NUMBER','ORIGIN','ORIGIN_CITY','DEST','DEST_CITY','DEP_TIME','DEP_DELAY','ARR_DELAY','DISTANCE','DIVERTED','ori_TMIN','ori_TMAX','ori_SNOW','ori_SNWD','ori_AWND','ori_PRCP','dest_TMIN','dest_TMAX','dest_SNOW','dest_SNWD','dest_AWND','dest_PRCP','DELAYED']

# Drop all columns except the specified ones
flight_weather_final= filtered_data.drop(columns=filtered_data.columns.difference(columns_to_keep_two))


##export the new table##
flight_weather_final.to_csv('flight_weather_final.csv', index=False)

flight_weather_final.info()
```

```python
flight_weather_final.info()
```

```python
##change date in  to datetime object
flight_weather_final["FL_DATE"] = pd.to_datetime(flight_weather_final["FL_DATE"])

flight_weather_final['month'] = flight_weather_final['FL_DATE'].dt.month
flight_weather_final['is_weekend'] = flight_weather_final['FL_DATE'].dt.dayofweek >= 4  

flight_weather_final["month"] = pd.Categorical(flight_weather_final["month"])
flight_weather_final.info()

```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer ##how we handle missing values
from sklearn.compose import ColumnTransformer ##Transform our data features
```

```python
#extract Numeric Features
numeric_features = flight_weather_final.select_dtypes(
    include=["float","int"]
).drop(columns=["DEP_TIME",'FL_NUMBER','DEP_DELAY','ARR_DELAY','DIVERTED','FL_NUMBER'])

print(numeric_features.head(2))
categorical_features = flight_weather_final.iloc[:,~flight_weather_final.columns.isin(numeric_features.columns.values)]

#one-hot encoding
categorical_features= flight_weather_final.select_dtypes(include=["category", "object",'bool'])
categorical_features.head(2)
```

```python
from sklearn.model_selection import train_test_split
X = flight_weather_final.iloc[:,~flight_weather_final.columns.isin(["DELAYED"])]
y = flight_weather_final["DELAYED"]

X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    shuffle=True,
    random_state=777,
    test_size=0.15
)
X_train.head(2)
```

```python
numeric_columns =numeric_features.drop(columns=["DELAYED"]).columns.values
categorical_columns = categorical_features.columns.values
numeric_pipe = Pipeline(
    steps=[
        ("imputer", SimpleImputer()),
        ("standardization", StandardScaler())

    ]
)
```

```python
categorical_pipe = Pipeline(
    steps=[
        ("one-hot-encoding", OneHotEncoder(handle_unknown="ignore"))
    ]
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression

preprocessing_pipe = ColumnTransformer(
    [
        ("numeric_pipe",numeric_pipe,numeric_columns),
        ("categorical_pipe", categorical_pipe, categorical_columns)
    ]
)
model = Pipeline(
    [("Preprocessing",preprocessing_pipe),
     ("regression_model", LogisticRegression())

    ]
    
)
```

```python
from sklearn.metrics import accuracy_score, classification_report

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

#Evaluate
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(accuracy)
print(report)
```
