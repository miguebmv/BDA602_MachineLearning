```python
import pandas as pd

df = pd.read_csv("filtered_flight_weather.csv")
df.head()
```

```python
df[df['ORIGIN_CITY']=='San Diego, CA']
```

```python
columns_keep = ["FL_DATE","AIRLINE","ORIGIN","ORIGIN_CITY","DEST","DEST_CITY","CRS_DEP_TIME","DEP_DELAY","CRS_ARR_TIME","ARR_DELAY","CANCELLED",
                "DIVERTED","DISTANCE","ori_TMIN","ori_TMAX","ori_SNOW","ori_SNWD","ori_AWND","ori_PRCP","dest_TMIN","dest_TMAX","dest_SNOW",
                "dest_SNWD","dest_AWND","dest_PRCP"]
sub_df = df[columns_keep]
sub_df.head()

```

```python
new_df = sub_df
new_df['CRS_DEP_TIME'] = sub_df['CRS_DEP_TIME'].astype(str)

new_df['CRS_DEP_TIME'] = sub_df['CRS_DEP_TIME'].str.replace(r'\.0$', '', regex=True)

new_df['CRS_DEP_TIME'] = new_df['CRS_DEP_TIME'].str.zfill(4)

new_df.head()

#df['CRS_DEP_TIME'] = df['CRS_DEP_TIME'].str.zfill(4)
```

```python
new_df['CRS_DEP_TIME'] = pd.to_datetime(new_df['CRS_DEP_TIME'], format='%H%M')
new_df['FL_DATE'] = pd.to_datetime(new_df['FL_DATE'])

new_df.info()
```

```python
DEP_time_timedelta = new_df['CRS_DEP_TIME'] - new_df['CRS_DEP_TIME'].dt.normalize()
new_df['DEP_datetime'] = new_df['FL_DATE'] + DEP_time_timedelta


new_df.head()
```

```python
news_df = new_df.drop(columns=["FL_DATE","CRS_DEP_TIME"])
news_df.rename(columns={'DEP_datetime':'CRS_DEP_Datetime'},inplace=True)
news_df.head()
```

```python
## Add a delayed column
condition = (news_df['ARR_DELAY'] >= 15) | (news_df['DEP_DELAY'] >= 15)

# Create DELAYED column based on condition
news_df['15_DELAYED'] = condition.astype(int)

news_df.head()
```

```python
#make categorical vars for datetime

news_df['year'] = news_df['CRS_DEP_Datetime'].dt.year
news_df['month'] = news_df['CRS_DEP_Datetime'].dt.month
news_df['day'] = news_df['CRS_DEP_Datetime'].dt.day
news_df['day_of_week'] = news_df['CRS_DEP_Datetime'].dt.dayofweek  # Monday=0, Sunday=6
news_df['hour'] = news_df['CRS_DEP_Datetime'].dt.hour

news_df.head()
```

```python
news_df.to_csv("rileys_data.csv",index=False)
```
